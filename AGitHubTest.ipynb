Setting Up the Environment for Data Analysis and Visualization

```python
# Data manipulation and analysis
import pandas as pd
import numpy as np

# Data visualization
import matplotlib.pyplot as plt
import seaborn as sns

# Warnings (to ignore unnecessary warnings)
import warnings
warnings.filterwarnings("ignore")

# Display settings
pd.set_option('display.max_columns', None)  # Show all columns in the output
plt.style.use('ggplot')  # Set a consistent plot style
sns.set_palette('muted')  # Set seaborn palette
```

Data Overview and Initial Assessment

```python

# Load the dataset with specified encoding
file_path = "/content/Superstore-Data.csv"
df = pd.read_csv(file_path, encoding='ISO-8859-1')

# Display first few rows of the dataset
print("First 5 rows of the dataset:")
display(df.head())

# Display basic information about the dataset
print("\nDataset Information:")
df.info()

# Statistical summary for numerical columns
print("\nStatistical Summary (Numerical Columns):")
display(df.describe())

# Check for duplicate rows
duplicates = df.duplicated().sum()
print(f"\nNumber of duplicate rows: {duplicates}")

# Check for missing values
missing_values = df.isnull().sum()
print("\nMissing Values in Each Column:")
display(missing_values)

# Preview column names
print("\nColumn Names:")
print(df.columns.tolist())

# Preview dataset shape
print(f"\nDataset Shape: {df.shape}")
```

Statistical Analysis

```python
import scipy.stats as stats
import pandas as pd
import numpy as np

# 1. Profit by Year - ANOVA Test (to check if profits differ significantly across years)
profit_by_year = df.groupby('Year')['Profit'].apply(list)
anova_year = stats.f_oneway(*profit_by_year)
print("Profit by Year - ANOVA Test")
print(f"F-statistic: {anova_year.statistic:.2f}, p-value: {anova_year.pvalue:.4f}")
print("Significant difference" if anova_year.pvalue < 0.05 else "No significant difference")
print("-" * 50)

# 2. Profit by City - Kruskal-Wallis Test (non-parametric test for profit distribution across top 10 cities)
top_cities = df.groupby('City')['Profit'].sum().nlargest(10).index
profit_by_city = [df[df['City'] == city]['Profit'].values for city in top_cities]
kruskal_city = stats.kruskal(*profit_by_city)
print("Profit by City (Top 10) - Kruskal-Wallis Test")
print(f"H-statistic: {kruskal_city.statistic:.2f}, p-value: {kruskal_city.pvalue:.4f}")
print("Significant difference" if kruskal_city.pvalue < 0.05 else "No significant difference")
print("-" * 50)

# 3. Profit by Region - ANOVA Test (to check if profits differ significantly across regions)
profit_by_region = df.groupby('Region')['Profit'].apply(list)
anova_region = stats.f_oneway(*profit_by_region)
print("Profit by Region - ANOVA Test")
print(f"F-statistic: {anova_region.statistic:.2f}, p-value: {anova_region.pvalue:.4f}")
print("Significant difference" if anova_region.pvalue < 0.05 else "No significant difference")
print("-" * 50)

# 4. Profit by Quarter - ANOVA Test (to check if profits differ significantly across quarters)
profit_by_quarter = df.groupby('Quarter')['Profit'].apply(list)
anova_quarter = stats.f_oneway(*profit_by_quarter)
print("Profit by Quarter - ANOVA Test")
print(f"F-statistic: {anova_quarter.statistic:.2f}, p-value: {anova_quarter.pvalue:.4f}")
print("Significant difference" if anova_quarter.pvalue < 0.05 else "No significant difference")
print("-" * 50)

# 5. Profit by Customer Name - Variance Test (Levene’s Test to check if profit variance is significant)
top_customers = df.groupby('Customer Name')['Profit'].sum().nlargest(10).index
profit_by_customer = [df[df['Customer Name'] == customer]['Profit'].values for customer in top_customers]
levene_customer = stats.levene(*profit_by_customer)
print("Profit by Customer Name (Top 10) - Levene's Test")
print(f"W-statistic: {levene_customer.statistic:.2f}, p-value: {levene_customer.pvalue:.4f}")
print("Significant variance difference" if levene_customer.pvalue < 0.05 else "No significant variance difference")
print("-" * 50)

# 6. Relationship between Profit and Processing Time - Pearson Correlation
df.dropna(subset=['Processing Time'], inplace=True)  # Drop rows with missing processing time
pearson_corr, p_value = stats.pearsonr(df['Processing Time'], df['Profit'])
print("Relationship between Profit and Processing Time - Pearson Correlation")
print(f"Correlation Coefficient: {pearson_corr:.2f}, p-value: {p_value:.4f}")
print("Significant correlation" if p_value < 0.05 else "No significant correlation")
print("-" * 50)
```
